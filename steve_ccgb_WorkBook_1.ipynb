```python
def newCrop_entry(veg_or_frut_name, family):
    
    if not Crop.objects.filter(name=veg_or_frut_name).exists():
        new_crop_entry = Crop(
                        business_id = 1,
                        name = veg_or_frut_name,
                        age_in_days = 100,
                        notes = veg_or_frut_name
                        )
        new_crop_entry.save()
        
    else:
        new_crop_entry = Crop.objects.get(name = veg_or_frut_name)

    if not CropGroup.objects.filter(name=family).exists():
        new_cropgroup_entry= CropGroup(
                            business_id = 1,
                            name = family,
                            short_name = family
                            )
        new_cropgroup_entry.save()
    
    else:
        new_cropgroup_entry = CropGroup.objects.get(name=family)
        
    if not CropCropGroupMap.objects.filter(crop_id=new_crop_entry.id, crop_group_id = new_cropgroup_entry.id):
        new_CropCropGroupMap = CropCropGroupMap(   
                            crop_id = new_crop_entry.id,
                            crop_group_id = new_cropgroup_entry.id
                            )
        new_CropCropGroupMap.save()
        print('mapped')
    else:
        print('already exists')

    
```
