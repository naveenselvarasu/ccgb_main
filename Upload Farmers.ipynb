```python
# Uploading agents into Farmer model

business = Business.objects.get(id=1)

count = 0
for index, row in df.iterrows():
    count = count+1
    
    print('{}:{}'.format(row['Farmer  First Name'], row['New Farmer Code']))
    
    
    # Add new hobli if not available already
    hobli, created = Hobli.objects.get_or_create(
        name=str(row['Hobli']),
        taluk=Taluk.objects.get(name=row['Taluk']),
        defaults = {
        }
    )
    if created:
        print("\t\t", hobli.name, " Hobli created")
    else:
        print("\t\t", hobli.name, " Hobli existed already")
        
    if 'xxx' in row['Hobli']:
        if Hobli.objects.filter(taluk__name=row['Taluk']).exists():
            hobli = Hobli.objects.filter(taluk__name=row['Taluk'])[0]
            
        else:
            hobli = Hobli.objects.get(name='xxx', taluk__name='xxx')
           

    # Add new hobli if not available already
    village_name=row['Village'].rstrip(' ')
    village, created = Village.objects.get_or_create(
        name=str(village_name),
        hobli=Hobli.objects.get(name=row['Hobli'], taluk=Taluk.objects.get(name=row['Taluk'])),
        defaults = {
        }
    )
    if created:
        print("\t\t", village.name, "\tVillage created")
    else:
        print("\t\t", village.name, "\tVillage existed already")
    
    
    # Get Position User, Agent User
    position_user = User.objects.get(id=PositionPositionUserMap.objects.get(position__name='AFS01').user_id)
    agent_user = User.objects.get(id=UserProfile.objects.get(code=row['Agent Code']).user_id)
    cluster = Cluster.objects.get(name=row['Cluster'])
    
    
    farmer, created = Farmer.objects.get_or_create(
        code=str(row['New Farmer Code']),
        first_name=row['Farmer  First Name'],
        last_name=row['Farmer Last  Name'],
        business=business,
        
        defaults = {
            'gender': Gender.objects.get(name='Male'),
            'state':State.objects.get(name='Karnataka'),
            'district':District.objects.get(name=row['District']),
            'taluk':Taluk.objects.get(name=row['Taluk']),
            'hobli':hobli,
            'village':village,
            'communication_language':Language.objects.get(name='Kannada'),
            'cluster':cluster,
            'farmer_data_source':farmer_data_source,
            'created_by':User.objects.get(id=1),
            'modified_by':User.objects.get(id=1)
        }
    )
    if created:
        print("\t\t", farmer.code,  " Farmer created")
    else:
        print("\t\t", farmer.code,  " Farmer existed already")
        
    
    
    # Link farmers with afs
    afsfarmermap, created = AfsFarmerMap.objects.get_or_create(
        season = Season.objects.get(name='2019-Marigold-01'),
        officer = position_user,
        farmer = farmer,
        defaults = {
        }
    )
    if created:
        print("\t\t", afsfarmermap.officer, "-", afsfarmermap.farmer.code, "\t afsfarmermap created")
    else:
        print("\t\t", afsfarmermap.officer, "-", afsfarmermap.farmer.code, "\t afsfarmermap existed already")
    
    
    # Link farmers with agents
    agentfarmermap, created = AgentFarmerMap.objects.get_or_create(
        season = Season.objects.get(name='2019-Marigold-01'),
        agent = agent_user,
        farmer = farmer,
        defaults = {
        }
    )
    agent_code = UserProfile.objects.get(user=agent_user).code
    if created:
        print("\t\t", agentfarmermap.agent.first_name, '(', agent_code, ')', "<->", agentfarmermap.farmer.code, "\t agentfarmermap created")
    else:
        print("\t\t", agentfarmermap.agent.first_name, '(', agent_code, ')', "<->", agentfarmermap.farmer.code, "\t agentfarmermap existed already")
    
    
```
