```python
for sowing_id in list(SowingBoundary.objects.filter().values_list('sowing__id',flat=True)):
    if not SowingBoundaryMap.objects.filter(sowing_id=sowing_id).exists():
        sowing_boundary_map = SowingBoundaryMap(
            sowing_id=sowing_id,
            data_source_id=1,
            added_by_id=1
        )
        sowing_boundary_map.save()    
        for boundary in SowingBoundary.objects.filter(sowing_id=sowing_id):
            sowing_boundary_map.sowing_boundry.add(boundary)
        sowing_boundary_map.save()    
        print('saved')
for row in SowingBoundaryMap.objects.filter():
    row.added_by = row.sowing.farmer.created_by
    row.save()
    print('added by added')
```
